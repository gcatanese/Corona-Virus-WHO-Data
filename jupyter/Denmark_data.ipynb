```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../data/csv/new_cases.csv'
df = pd.read_csv(filename)
```

```python
d_from = '2020-03-01'
d_to = '2020-04-15'

df = df[(df['date'] >= d_from) & (df['date'] <= d_to)]
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%d/%m')
```

```python
all_data = df[['date', 'Norway']]
all_data = all_data.rename(columns={'Norway': "New Cases"})
all_data.tail(3)
```

```python
filename = '../data/csv/new_deaths.csv'
df2 = pd.read_csv(filename)
```

```python

df2 = df2[(df2['date'] >= d_from) & (df2['date'] <= d_to)]
df2['date'] = pd.to_datetime(df2['date'])
df2['date'] = df2['date'].dt.strftime('%d/%m')

all_data2 = df2[['date', 'Denmark']]
all_data2 = all_data2.rename(columns={'Denmark': "New Deaths"})
all_data2.tail(3)
```

```python
# merging dataframes
df3 = pd.merge(all_data, all_data2, how='outer', on='date')
df3.tail(3)
```

```python
import matplotlib.dates as mdates
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

plt.style.use('Solarize_Light2')
plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'axes.labelsize': 18})
#plt.rcParams['axes.facecolor'] = 'lightslategray'
plt.rcParams.update({'xtick.labelsize': 'x-large'})
plt.rcParams.update({'ytick.labelsize': 'x-large'})


#['Solarize_Light2', '_classic_test_patch', 'bmh', 'classic', 'dark_background', 'fast', 'fivethirtyeight', 'ggplot', 
# 'grayscale', 'seaborn', 'seaborn-bright', 'seaborn-colorblind', 'seaborn-dark', 'seaborn-dark-palette', 'seaborn-darkgrid', 'seaborn-deep', 'seaborn-muted', 
#'seaborn-notebook', 'seaborn-paper', 'seaborn-pastel', 'seaborn-poster', 'seaborn-talk',  'seaborn-ticks', 'seaborn-white', 'seaborn-whitegrid', 'tableau-colorblind10']
#ax.set_axis_bgcolor("lightslategray")


ax = all_data.plot.bar(x='date', figsize=(20,10), width = 0.5, color='b')

ax.set_xlabel('')

ax.xaxis.set_major_locator(MultipleLocator(1))

ax.set_xlim(left=1)
#ax.set_ylim(bottom=0)

ax.set_title('Denmark - Daily New Cases', fontsize= 30)

ax.get_legend().remove()


plt.show()
```
