```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../data/csv/dpc-covid19-ita-regioni.csv'
df = pd.read_csv(filename)
```

```python
d_from = '2020-03-01'
d_to = '2020-03-31'

df = df[(df['data'] >= d_from) & (df['data'] <= d_to)]
df = df[['data', 'denominazione_regione', 'nuovi_attualmente_positivi']]
```

```python
REGIONS = ['Sicilia','Puglia','Lombardia','Veneto']

all_data = df[df['denominazione_regione'].isin(REGIONS)]
```

```python

sicilia_data = df[df['denominazione_regione'].isin(['Sicilia'])]
sicilia_data = sicilia_data.rename(columns={"nuovi_attualmente_positivi": "Sicilia"})

puglia_data = df[df['denominazione_regione'].isin(['Puglia'])]
puglia_data = puglia_data.rename(columns={"nuovi_attualmente_positivi": "Puglia"})

lombardia_data = df[df['denominazione_regione'].isin(['Lombardia'])]
lombardia_data = lombardia_data.rename(columns={"nuovi_attualmente_positivi": "Lombardia"})

veneto_data = df[df['denominazione_regione'].isin(['Veneto'])]
veneto_data = veneto_data.rename(columns={"nuovi_attualmente_positivi": "Veneto"})

```

```python
import matplotlib.dates as mdates
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.facecolor'] = 'whitesmoke'

ax = sicilia_data.plot(figsize=(20,10), x='data')
puglia_data.plot(ax=ax, x='data')
lombardia_data.plot(ax=ax, x='data')
veneto_data.plot(ax=ax, x='data')

ax.set_xlabel('Day')

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)


plt.show()
```
