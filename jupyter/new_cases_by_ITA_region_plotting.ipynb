```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../data/csv/dpc-covid19-ita-regioni.csv'
df = pd.read_csv(filename)
```

```python
d_from = '2020-03-01'
d_to = '2020-03-31'
#FIELD = 'totale_positivi'
FIELD = 'tamponi'

df = df[(df['data'] >= d_from) & (df['data'] <= d_to)]
df = df[['data', 'denominazione_regione', FIELD]]
```

```python
REGIONS = ['Sicilia','Puglia','Lombardia','Veneto']

all_data = df['denominazione_regione']
```

```python

sicilia_data = df[df['denominazione_regione'].isin(['Sicilia'])]
sicilia_data = sicilia_data.rename(columns={FIELD: "Sicilia"})

puglia_data = df[df['denominazione_regione'].isin(['Puglia'])]
puglia_data = puglia_data.rename(columns={FIELD: "Puglia"})

lombardia_data = df[df['denominazione_regione'].isin(['Lombardia'])]
lombardia_data = lombardia_data.rename(columns={FIELD: "Lombardia"})

veneto_data = df[df['denominazione_regione'].isin(['Veneto'])]
veneto_data = veneto_data.rename(columns={FIELD: "Veneto"})

lazio_data = df[df['denominazione_regione'].isin(['Lazio'])]
lazio_data = lazio_data.rename(columns={FIELD: "Lazio"})

sardegna_data = df[df['denominazione_regione'].isin(['Sardegna'])]
sardegna_data = sardegna_data.rename(columns={FIELD: "Sardegna"})

calabria_data = df[df['denominazione_regione'].isin(['Calabria'])]
calabria_data = calabria_data.rename(columns={FIELD: "Calabria"})

campania_data = df[df['denominazione_regione'].isin(['Campania'])]
campania_data = campania_data.rename(columns={FIELD: "Campania"})

piemonte_data = df[df['denominazione_regione'].isin(['Campania'])]
piemonte_data = piemonte_data.rename(columns={FIELD: "Piemonte"})
```

```python
import matplotlib.dates as mdates
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.facecolor'] = 'whitesmoke'

ax = lazio_data.plot(figsize=(20,10), x='data')
#veneto_data.plot(ax=ax, x='data')
sicilia_data.plot(ax=ax, x='data')
puglia_data.plot(ax=ax, x='data')
campania_data.plot(ax=ax, x='data')
calabria_data.plot(ax=ax, x='data')
sardegna_data.plot(ax=ax, x='data')

ax.set_xlabel('Day')

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
ax.set_xlim(left=0)
#ax.set_ylim(bottom=0)


plt.show()
```
