```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../data/csv/total_cases.csv'
df = pd.read_csv(filename)
```

```python
d_from = '2020-02-28'
d_to = '2020-03-18'

df = df[(df['date'] >= d_from) & (df['date'] <= d_to)]
df['date'] = pd.to_datetime(df['date'])
```

```python
all_data = df[['date', 'Italy', 'United Kingdom', 'Netherlands', 'Spain']]
```

```python
import datetime
ax = all_data.plot(figsize=(20,10), x='date')

plt.axvline(x= datetime.datetime(2020, 3, 9, 0, 0), color = 'blue')
plt.axvline(x= datetime.datetime(2020, 3, 16, 0, 0), color = 'green', label = 'NL Lockdown')
plt.axvline(x= datetime.datetime(2020, 3, 14, 0, 0), color = 'red', label = 'Spain Lockdown')
plt.legend()

left, width = .50, .5
bottom, height = .50, .5
right = left + width
top = bottom + height

ax.text(.50, bottom, 'Italy Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(left, bottom, 'Italy Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

plt.show()
```
