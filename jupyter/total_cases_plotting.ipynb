```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../data/csv/total_cases.csv'
df = pd.read_csv(filename)
```

```python
d_from = '2020-03-01'
d_to = '2020-03-31'

df = df[(df['date'] >= d_from) & (df['date'] <= d_to)]
df['date'] = pd.to_datetime(df['date'])
```

```python
all_data = df[['date', 'Italy', 'United Kingdom', 'Netherlands', 'Spain']]
all_data.head
```

```python
import datetime

plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.facecolor'] = 'whitesmoke'

ax = all_data.plot(figsize=(20,10), x='date', color=('b','r','orange','g'))


plt.axvline(x= datetime.datetime(2020, 3, 9, 0, 0), color = 'b')  # IT lockdown
plt.axvline(x= datetime.datetime(2020, 3, 21, 0, 0), color = 'r') # UK lockdown
plt.axvline(x= datetime.datetime(2020, 3, 16, 0, 0), color = 'orange') # NL lockdown
plt.axvline(x= datetime.datetime(2020, 3, 14, 0, 0), color = 'g') # ES lockdown
plt.legend()

left, width = .05, .5
bottom, height = .70, .5
right = left + width
top = bottom + height

#ax.xaxis.set_major_locator(MultipleLocator(1))


ax.text(.22, .70, 'Italy Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(.60, .60, 'NL Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(.67, .80, 'UK Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(.41, .80, 'ES Lockdown',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

plt.show()
```
